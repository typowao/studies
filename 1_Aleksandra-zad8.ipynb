```python
#wczytanie w zmienna danych z nmt
a=open(r"C:\Users\ola\Documents\pw\PROGRAMOWANIE W SIP\zajęcia_nr_1\1_pliki_tekstowe\nmt.txt")
   
#utworzenie pustej tablicy    
wynik=[]

#usuniecie enterów z koncow linijek i podzielenie ciagu na liste i dodanie do listy wynik
for line in a.readlines():
    b=line.strip().split()
    wynik.append(b)
a.close()

#zamiana krotek na liste
b=list(wynik)

#wyswietlenie listy tak, aby kazda liczba byla w nowym wierszu, aby w dalszym kroku bylo wiadomo, 
#ktory wiersz i ktora kolumne wybrac 
for i in range(len(b)):
    for j in range(len(b[i])):
        print(b[i][j]);
    #print();
    
#utworzenie listy, do ktorej beda zapisane tylko wysokosci
dane=[]
for i in range(len(b)):
    c=b[i][2];
    dane.append(c);
print(dane)
```

```python
#zamiana tekstu na liczby zmiennoprzecinkowe
dane_w_float = list(map(float, dane))

#obliczenie sredniej wysokosci
srednia_wys = statistics.mean(dane_w_float)
print(srednia_wys)
```
